```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Bird_Strikes.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
# @title Distribution of Wildlife Struck by Species

import pandas as pd

species_count = df['Wildlife: Species'].value_counts()
plt.figure(figsize=(10,6))
species_count.plot(kind='bar')
plt.title('Distribution of Wildlife Struck by Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# @title Number of Wildlife Strikes by Aircraft Type

aircraft_type_count = df['Aircraft: Type'].value_counts()
plt.figure(figsize=(10,6))
aircraft_type_count.plot(kind='bar', color='orange')
plt.title('Number of Wildlife Strikes by Aircraft Type')
plt.xlabel('Aircraft Type')
plt.ylabel('Count')
plt.show()
```

```python
# @title Impact to Flight Distribution

impact_count = df['Effect: Impact to flight'].value_counts()
plt.figure(figsize=(8,5))
impact_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='viridis')
plt.title('Impact to Flight Distribution')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# @title Number of Wildlife Strikes by Airport (Top 10)

airport_count = df['Airport: Name'].value_counts().head(10)  # Top 10 airports
plt.figure(figsize=(12,8))
airport_count.plot(kind='bar', color='purple')
plt.title('Number of Wildlife Strikes by Airport (Top 10)')
plt.xlabel('Airport')
plt.ylabel('Count')
plt.show()
```
